```python
data_clean = data_clean.rename(columns = {"mass (g)":"massg"})
```

```python
data_clean['mass'] = data_clean['massg']/1000
data_clean.head()
```

```python
data_clean['masslog'] = data_clean['mass'] + 1
#values under 1 inputed into log() create negative values
#since the mass (in kg) will be used to determine the circle's radius which must be positive 
#And the map will display relative size, not accurate size
#I am assigning all values under 1 to 1 
data_clean.head()
```

```python
data_clean['mass'] = data_clean['mass'].astype(float)
data_clean['massg'] = data_clean['massg'].astype(float)
data_clean['masslog'] = data_clean['masslog'].astype(float)
print(len(data_clean))
data_clean.head()

#getting rid of scientific notation and other anomalies 


```
